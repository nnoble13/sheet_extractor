```python
'''This is a script that will take an HTML version of a view-only Google
Sheets file and produce the equivalent and editable CSV file.'''

from bs4 import BeautifulSoup
import pandas as pd
```

```python
file_path = 'sheet.html'
with open(file_path, 'r') as file:
    soup = BeautifulSoup(file.read(), 'html.parser')
```

```python
table = soup.find('table', class_='waffle') # look for only the table
```

```python
table
```

```python
headers = []
for table_header in table.find_all('th')[1:]:
    headers.append(table_header.get_text(strip=True))
```

```python
rows = []
for tr in table.find_all('tr')[1:]:
    cells = []
    for td in tr.find_all('td'):
        cells.append(td.get_text(strip=True))
    rows.append(cells)
```

```python
max_columns = max(len(row) for row in rows)
```

```python
adjusted_rows = [row + [''] * (max_columns - len(row)) for row in rows] #why am i doing this?!
adjusted_headers = headers[:max_columns] if len(headers) >= max_columns else headers + [''] * (max_columns - len(headers))
```
